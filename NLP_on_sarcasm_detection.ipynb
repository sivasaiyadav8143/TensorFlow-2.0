```
import json
```

```
with open('/tmp/sarcasm.json','r') as fl:
  df = json.load(fl)
```

```
df[0]
```

```
article_link = []
headline = []
is_sarcastic = []

for i in df:
  article_link.append(i['article_link'])
  headline.append(i['headline'])
  is_sarcastic.append(i['is_sarcastic'])
```

```
headline[:3]
```

```
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```
token = Tokenizer(oov_token='OOV')
token.fit_on_texts(headline)
print(len(token.index_word))
print(token.index_word)
```

```
seq = token.texts_to_sequences(headline)
padd = pad_sequences(seq)
print(padd[0])
print(padd.shape)
```
